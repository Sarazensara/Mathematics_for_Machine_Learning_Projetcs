# Symmetric Matrix and Eigenvalues

### Problem:
Let $S$ be a symmetric matrix, i.e., $S = S^T$, and suppose that $S$ has orthonormal eigenvectors $v_1, v_2, \dots, v_n$. Any vector $x$ can be written as a linear combination of these eigenvectors:

$$
x = c_1 v_1 + c_2 v_2 + \cdots + c_n v_n
$$

### Explain the following formulas:

1. **First Formula:**
   $$
   x^T x = c_1^2 + c_2^2 + \cdots + c_n^2
   $$
   - What does this formula represent? How does it relate to the vector $x$ and its coefficients $c_1, c_2, \dots, c_n$?

2. **Second Formula:**
   $$
   x^T S x = \lambda_1 c_1^2 + \lambda_2 c_2^2 + \cdots + \lambda_n c_n^2
   $$
   - What does this formula represent? How does it connect to the eigenvalues $\lambda_1, \lambda_2, \dots, \lambda_n$ of the matrix $S$?

---

## **Solution:**

### **Proof:**
Since $S$ is symmetric, we know that it has an orthonormal set of eigenvectors $\{ v_1, v_2, \dots, v_n \}$. This means any vector $x$ in this space can be written as:

$$
x = c_1 v_1 + c_2 v_2 + \cdots + c_n v_n
$$

where $c_i = v_i^T x$ are the coefficients when projecting $x$ onto the eigenvectors. 

#### **First Formula:**
To compute $x^T x$, we use the property that eigenvectors are orthonormal:

$$
x^T x = (c_1 v_1 + c_2 v_2 + \dots + c_n v_n)^T (c_1 v_1 + c_2 v_2 + \dots + c_n v_n)
$$

Expanding this and using the fact that $v_i^T v_j = 1$ if $i = j$ and $0$ otherwise:

$$
x^T x = c_1^2 v_1^T v_1 + c_2^2 v_2^T v_2 + \dots + c_n^2 v_n^T v_n
$$

Since $v_i^T v_i = 1$, we simplify to:

$$
x^T x = c_1^2 + c_2^2 + \dots + c_n^2
$$

This shows that the squared length (norm) of $x$ is simply the sum of the squares of its coefficients.

#### **Second Formula:**
Now, consider the quadratic form $x^T S x$:

$$
x^T S x = (c_1 v_1 + c_2 v_2 + \dots + c_n v_n)^T S (c_1 v_1 + c_2 v_2 + \dots + c_n v_n)
$$

Since each eigenvector satisfies $S v_i = \lambda_i v_i$, we substitute this:

$$
x^T S x = (c_1 v_1 + c_2 v_2 + \dots + c_n v_n)^T (\lambda_1 c_1 v_1 + \lambda_2 c_2 v_2 + \dots + \lambda_n c_n v_n)
$$

Expanding and using the orthonormality of eigenvectors:

$$
x^T S x = \lambda_1 c_1^2 + \lambda_2 c_2^2 + \dots + \lambda_n c_n^2
$$

This formula shows that $x^T S x$ is a weighted sum of the squared coefficients, where each weight is the corresponding eigenvalue of $S$.

---

✅ **Conclusion:**  
- The first formula tells us how the squared length of $x$ is just the sum of squared coefficients.  
- The second formula shows how $x^T S x$ is a sum of these same squared coefficients but weighted by the eigenvalues of $S$.  

$$
\blacksquare
$$


# Rayleigh Quotient and Eigenvalues

### Problem:
Consider a matrix \( A \) and let \( x \) be a nonzero vector. The Rayleigh quotient is given by:

$$
R(A) = \frac{x^T A x}{x^T x}
$$

Use the Rayleigh quotient formula to explain why it is bounded above and below by the largest and smallest eigenvalues of \( A \), respectively.

---

## **Solution:**

### **Proof:**

Let \( A \) be a symmetric matrix with eigenvalues \( \lambda_1, \lambda_2, \dots, \lambda_n \) and corresponding orthonormal eigenvectors \( v_1, v_2, \dots, v_n \). Any vector \( x \) can be written as a linear combination of these eigenvectors:

$$
x = c_1 v_1 + c_2 v_2 + \cdots + c_n v_n
$$

where \( c_i = v_i^T x \) represents the projection of \( x \) onto the eigenvector \( v_i \).

#### **Step 1: Expression for \( x^T A x \)**

Since \( A \) is symmetric, we have \( A v_i = \lambda_i v_i \) for each eigenvector \( v_i \). Now, consider the quadratic form \( x^T A x \):

$$
x^T A x = (c_1 v_1 + c_2 v_2 + \dots + c_n v_n)^T A (c_1 v_1 + c_2 v_2 + \dots + c_n v_n)
$$

Expanding this expression:

$$
x^T A x = \sum_{i=1}^{n} c_i^2 \lambda_i
$$

Thus, \( x^T A x \) is a weighted sum of the eigenvalues, with each coefficient \( c_i^2 \) representing the squared projection of \( x \) onto the eigenvector \( v_i \).

#### **Step 2: Expression for \( x^T x \)**

The expression for \( x^T x \) is:

$$
x^T x = \sum_{i=1}^{n} c_i^2
$$

This is the sum of the squares of the coefficients \( c_1, c_2, \dots, c_n \), which represents the squared length (norm) of the vector \( x \).

#### **Step 3: Rayleigh Quotient Bound**

Now, consider the Rayleigh quotient:

$$
R(A) = \frac{x^T A x}{x^T x} = \frac{\sum_{i=1}^{n} c_i^2 \lambda_i}{\sum_{i=1}^{n} c_i^2}
$$

This is a weighted average of the eigenvalues, where the weights are the squared coefficients \( c_i^2 \).

Since \( \lambda_{\text{min}} \) and \( \lambda_{\text{max}} \) represent the smallest and largest eigenvalues of \( A \), respectively, the Rayleigh quotient satisfies the following bounds:

$$
\lambda_{\text{min}} \leq R(A) \leq \lambda_{\text{max}}
$$

This is because the Rayleigh quotient is a weighted average, and the smallest eigenvalue provides the lower bound while the largest eigenvalue provides the upper bound.

---

✅ **Conclusion:**  
The Rayleigh quotient \( R(A) = \frac{x^T A x}{x^T x} \) is bounded above and below by the largest and smallest eigenvalues of \( A \), respectively.

$$
\blacksquare
$$

# Eigenvectors of \( A A^T \)

### Problem:
Let \( A^T A v = \lambda v \) for some \( \lambda \neq 0 \). Find an eigenvector for \( A A^T \).

---

## **Solution:**

### **Proof:**

Given that \( A^T A v = \lambda v \) for some \( v \neq 0 \) and \( \lambda \neq 0 \), we aim to find an eigenvector of the matrix \( A A^T \).

#### **Step 1: Start with the given equation**

We know that:

$$
A^T A v = \lambda v
$$

This equation tells us that \( v \) is an eigenvector of the matrix \( A^T A \) with eigenvalue \( \lambda \).

#### **Step 2: Multiply both sides by \( A \)**

Now, apply \( A \) to both sides of the equation:

$$
A (A^T A v) = A (\lambda v)
$$

Since matrix multiplication is associative, we can rewrite the left-hand side:

$$
(A A^T) (A v) = \lambda (A v)
$$

#### **Step 3: Define a new vector**

Let \( w = A v \). Then the equation becomes:

$$
A A^T w = \lambda w
$$

This shows that \( w = A v \) is an eigenvector of \( A A^T \) with eigenvalue \( \lambda \).

#### **Step 4: Conclusion**

Therefore, \( A v \) is an eigenvector of \( A A^T \) corresponding to the eigenvalue \( \lambda \).

---

✅ **Conclusion:**  
If \( A^T A v = \lambda v \) for some \( \lambda \neq 0 \), then \( A v \) is an eigenvector of \( A A^T \) with the same eigenvalue \( \lambda \).

$$
\blacksquare
$$

## **Solution:**

### **Step 1: Singular Value Decomposition (SVD)**

The singular value decomposition (SVD) of a matrix \( A \) is given by:

$$
A = U \Sigma V^T
$$

where:
- \( U \) is a matrix of the left singular vectors (an orthogonal matrix),
- \( \Sigma \) is a diagonal matrix of singular values \( \sigma_i \),
- \( V \) is a matrix of the right singular vectors (an orthogonal matrix).

### **Matrix 1:**

#### Given matrix:
$$
A = \begin{pmatrix} 
3 & 4 \\

\end{pmatrix}
$$

#### **Step 1.1: Compute \( A^T A \) and its eigenvalues**

First, compute \( A^T A \):

$$
A^T A = \begin{pmatrix} 3 & 0 \\ 4 & 5 \end{pmatrix} \begin{pmatrix} 3 & 4 \\ 0 & 5 \end{pmatrix} = \begin{pmatrix} 9 & 15 \\ 15 & 41 \end{pmatrix}
$$

Next, find the eigenvalues of \( A^T A \) by solving the characteristic equation \( \det(A^T A - \lambda I) = 0 \):

$$
\det \begin{pmatrix} 9 - \lambda & 15 \\ 15 & 41 - \lambda \end{pmatrix} = 0
$$

Solving for \( \lambda \), we get the eigenvalues \( \lambda_1 = 49 \) and \( \lambda_2 = 1 \).

The singular values are the square roots of the eigenvalues, so:

$$
\sigma_1 = \sqrt{49} = 7, \quad \sigma_2 = \sqrt{1} = 1
$$

#### **Step 1.2: Find the right singular vectors \( v_1 \) and \( v_2 \)**

The right singular vectors correspond to the eigenvectors of \( A^T A \). Solving for the eigenvectors, we get:

$$
v_1 = \frac{1}{\sqrt{2}} \begin{pmatrix} 3 \\ 4 \end{pmatrix}, \quad v_2 = \frac{1}{\sqrt{2}} \begin{pmatrix} -4 \\ 3 \end{pmatrix}
$$

#### **Step 1.3: Find the left singular vectors \( u_1 \) and \( u_2 \)**

The left singular vectors correspond to the eigenvectors of \( A A^T \). Compute \( A A^T \):

$$
A A^T = \begin{pmatrix} 3 & 4 \\ 0 & 5 \end{pmatrix} \begin{pmatrix} 3 & 0 \\ 4 & 5 \end{pmatrix} = \begin{pmatrix} 25 & 15 \\ 15 & 25 \end{pmatrix}
$$

Solving for the eigenvectors of \( A A^T \), we find:

$$
u_1 = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 \\ 1 \end{pmatrix}, \quad u_2 = \frac{1}{\sqrt{2}} \begin{pmatrix} -1 \\ 1 \end{pmatrix}
$$

#### **Step 1.4: Assemble the SVD**

Thus, the SVD of \( A \) is:

# Singular Value Decomposition (SVD) and Inverse Matrices

### Problem:
Let \( A \) be an invertible matrix. Answer the following questions:

#### (a) Suppose you have the singular value decomposition of \( A \). How can you use it to find \( A^{-1} \)?

#### (b) Find all singular values of \( A^T A \).

---

## **Solution:**

### **Part (a): How to Find \( A^{-1} \) Using SVD**

The singular value decomposition (SVD) of a matrix \( A \) is given by:

$$
A = U \Sigma V^T
$$

where:
- \( U \) is a matrix of the left singular vectors (an orthogonal matrix),
- \( \Sigma \) is a diagonal matrix of the singular values \( \sigma_i \),
- \( V \) is a matrix of the right singular vectors (an orthogonal matrix).

To find the inverse of \( A \), we can use the following properties of the SVD:

1. **Inverse of \( U \) and \( V \)**:
   Since \( U \) and \( V \) are orthogonal matrices, their inverses are simply their transposes:
   $$ U^{-1} = U^T \quad \text{and} \quad V^{-1} = V^T $$

2. **Inverse of \( \Sigma \)**:
   The matrix \( \Sigma \) is diagonal, so its inverse is easy to compute. The inverse of a diagonal matrix is obtained by taking the reciprocal of each nonzero singular value:
   $$ \Sigma^{-1} = \text{diag}\left( \frac{1}{\sigma_1}, \frac{1}{\sigma_2}, \dots \right) $$

Thus, using the SVD, the inverse of \( A \) is given by:

$$
A^{-1} = V \Sigma^{-1} U^T
$$

This formula provides a straightforward way to compute the inverse of \( A \) using its SVD components.

---

### **Part (b): Singular Values of \( A^T A \)**

If \( A \) has the singular value decomposition:

$$
A = U \Sigma V^T
$$

then \( A^T \) has the decomposition:

$$
A^T = V \Sigma^T U^T
$$

Now, consider the matrix \( A^T A \):

$$
A^T A = (V \Sigma^T U^T)(U \Sigma V^T) = V \Sigma^T \Sigma V^T
$$

Since \( V \) is an orthogonal matrix, \( V^T V = I \), the identity matrix. Therefore:

$$
A^T A = V \Sigma^T \Sigma V^T
$$

The matrix \( \Sigma^T \Sigma \) is a diagonal matrix, and its diagonal entries are the squares of the singular values of \( A \), i.e., \( \sigma_i^2 \). Therefore, the singular values of \( A^T A \) are the squares of the singular values of \( A \).

Thus, the singular values of \( A^T A \) are:

$$
\sigma_1^2, \sigma_2^2, \dots
$$

These are simply the squared singular values of \( A \).

---

✅ **Conclusion:**

- In part (a), we showed how to find \( A^{-1} \) using the SVD of \( A \), with the formula \( A^{-1} = V \Sigma^{-1} U^T \).
- In part (b), we found that the singular values of \( A^T A \) are the squares of the singular values of \( A \).

$$
\blacksquare
$$
